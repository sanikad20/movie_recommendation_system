```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from scipy.sparse import csr_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies=pd.read_csv('/content/imdb_movies.csv')
```

```python
movies.head(5)
```

```python
movies.isnull().sum()
```

```python
movies['genre']=movies['genre'].fillna('')
movies['crew']=movies['crew'].fillna('')
movies['overview']=movies['overview'].fillna('')
```

```python
movies.isnull().sum()
```

```python
movies['combined_features']=movies['genre']+'.'+movies['overview']
```

```python
movies['combined_features'] = movies['combined_features'].fillna('')
```

```python
movies.head(5)
```

```python
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(movies['combined_features'])

```

```python
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
def recommend_movies(movie_title,num_recommendations=5):
  if movie_title not in movies['names'].values:
    return f"Movie '{movie_title}' not found in dataset."
  movie_idx = movies[movies['names'] == movie_title].index[0]
  similarity_scores = list(enumerate(cosine_sim[movie_idx]))
  similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
  similar_movies = similarity_scores[1:num_recommendations+1]
  recommended_movie_names = [movies.iloc[i[0]]['names'] for i in similar_movies]
  return recommended_movie_names
```

```python
recommendations = recommend_movies("Avatar: The Way of Water", num_recommendations=5)
print(recommendations)
```
